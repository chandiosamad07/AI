```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.pakwheels.com/used-cars/search/-/'
```

```python
req = requests.get(url)
```

```python
req.status_code
```

```python
soup = BeautifulSoup(req.text,'html.parser')
```

```python
Car_names = soup.find_all('a',class_='car-name')
```

```python
Car_list = []
```

```python
for car in Car_names:
    car = car.text.strip()
    Car_list.append(car)
```

```python
Car_list
```

```python
car_div = soup.find_all('div',class_="col-md-9 grid-style")
```

```python
cars = []
```

```python
for car_div in soup.find_all("div", class_="col-md-9 grid-style"):
    car_data = {}
    name_tag = car_div.find("a", class_="car-name")
    car_data["Name"] = name_tag.get_text(strip=True)
        
    city_tag = car_div.find("ul", class_="search-vehicle-info")
    car_data["City"] = city_tag.find("li").get_text(strip=True)
    
    price_tag = car_div.find("div", class_="price-details")
    car_data["Price"] = price_tag.get_text(strip=True)
    
    info_tags = car_div.find("ul", class_="search-vehicle-info-2").find_all("li")
    car_data["Year"] = info_tags[0].get_text(strip=True)
    car_data["Mileage"] = info_tags[1].get_text(strip=True)
    car_data["Fuel Type"] = info_tags[2].get_text(strip=True)
    car_data["Engine Capacity"] = info_tags[3].get_text(strip=True)
    car_data["Transmission"] = info_tags[4].get_text(strip=True)
    updated_tag = car_div.find("div", class_="pull-left dated")
    car_data["Last Updated"] = updated_tag.get_text(strip=True)
 
    cars.append(car_data)

for car in cars:
    print(car)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cars)
```

```python
df.to_csv("PakWheel_Used Car.csv")
```
