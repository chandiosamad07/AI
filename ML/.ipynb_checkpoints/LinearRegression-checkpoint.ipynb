Simple Linear Regression Model 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\abdul.samad\Downloads\ML\insurance.csv')
df.head()
```

```python
x_axis =df['Age']
```

```python
y_axis=df['Permium']
```

```python
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,y_axis)
plt.show()
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(df[['Age']],df[['Permium']])
```

```python
model.predict([[21]])
```

```python
model.predict([[20]])
```

```python
model.predict([[50]])
```

```python
model.coef_
```

```python
model.intercept_
```

Linear Regression Multivariable

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\abdul.samad\Downloads\ML\insurance-Multivariable.csv')
df.head()
```

```python
df.info()
```

```python
df['height'] = df['height'].fillna(df['height'].mean())
```

```python
df
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(df[['Age','height','weight']],df[['Permium']])
```

```python
model.predict([[27,167.56,60]])
```
